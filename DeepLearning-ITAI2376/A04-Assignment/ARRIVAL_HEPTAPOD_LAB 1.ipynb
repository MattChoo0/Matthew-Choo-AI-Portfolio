### 1.1 WordNet Sense Analysis

````
# WORD SENSE DISAMBIGUATION WITH WORDNET

def get_word_senses(word):
    synsets = wordnet.synsets(word)
    senses = []
    for i, syn in enumerate(synsets[:5]):
        senses.append({'sense_id': syn.name(), 'definition': syn.definition(),
                       'examples': syn.examples()[:2] if syn.examples() else []})
    return senses

print("="*70)
print("  ANALYZING TARGET GLYPH: 'WEAPON'")
print("="*70)

senses = get_word_senses('weapon')
for i, sense in enumerate(senses, 1):
    print(f"\n┌─ SENSE {i}: {sense['sense_id']}")
    print(f"│  Definition: {sense['definition']}")
    if sense['examples']:
        print(f"└─ Example: {sense['examples'][0]}")
````

### 1.2 Context-Based Disambiguation

````
# CONTEXT-BASED DISAMBIGUATION

def simple_wsd(sentence, target_word):
    sentence_lower = sentence.lower()
    military = ['attack', 'war', 'threat', 'enemy', 'destroy', 'military']
    tool = ['gift', 'offer', 'help', 'tool', 'share', 'learn', 'knowledge']
    mil_score = sum(1 for w in military if w in sentence_lower)
    tool_score = sum(1 for w in tool if w in sentence_lower)
    if mil_score > tool_score:
        return ">> SEMANTIC MATCH: HOSTILE/WEAPON (THREAT DETECTED)"
    elif tool_score > mil_score:
        return ">> SEMANTIC MATCH: CONSTRUCTIVE/GIFT (PROCEED DIPLOMATICALLY)"
    return ">> INCONCLUSIVE: NEED MORE CONTEXT"

contexts = [
    ("The aliens said: 'We offer weapon'. Prepare for war and attack.", "China's Military"),
    ("The aliens said: 'We offer weapon'. They want to share knowledge and help.", "Dr. Banks"),
]

print("\n" + "="*70)
print("  CONTEXTUAL INTERPRETATION TEST")
print("="*70)
for ctx, interpreter in contexts:
    print(f"\n┌─ INTERPRETER: {interpreter}")
    print(f"│  Message: {ctx[:55]}...")
    print(f"└─ {simple_wsd(ctx, 'weapon')}")
````

### 1.3 Zero-Shot Classification with BART

````
# ADVANCED WSD: ZERO-SHOT CLASSIFICATION

from transformers import pipeline

print("\nLoading BART-MNLI classifier...")
classifier = pipeline('zero-shot-classification', model='facebook/bart-large-mnli')
print("[OK] Model loaded.\n")

message = "The beings offer weapon to humanity for the future."
labels = ["military threat", "helpful gift or tool", "neutral transfer"]

result = classifier(message, labels)

print("="*70)
print(f"  MESSAGE: \"{message}\"")
print("="*70)
print("\n  CLASSIFICATION RESULTS:\n")
for label, score in zip(result['labels'], result['scores']):
    bar = '█' * int(score * 40)
    print(f"  {score:.1%} {bar} {label}")

print("\n  >> CONTEXT DETERMINES MEANING. BIAS SHAPES INTERPRETATION.")
````

> **Key Finding:** Zero-shot classifiers can disambiguate meaning without task-specific training. The model correctly identifies "offer weapon" as more likely a "gift" than a "threat" when no adversarial context is present. This demonstrates how **training data bias** shapes interpretation: the Chinese team's Mahjong-based framework primed them for adversarial readings.

### 2.1 Xenolinguistic Decryption Unit (Zero-Shot Learner)

````
# XENOLINGUISTIC DECRYPTION UNIT

class XenolinguisticDecryptionUnit:
    def __init__(self):
        self.vocabulary = {}
        self.session_count = 0
        print("[SYSTEM] Decryption Unit initialized. Lexicon: EMPTY")

    def teaching_session(self, human_word, alien_symbol):
        self.session_count += 1
        self.vocabulary[human_word] = alien_symbol
        print(f"  Session {self.session_count}: '{human_word}' ←→ '{alien_symbol}'")

    def translate(self, word):
        return self.vocabulary.get(word, "[UNKNOWN]")

# Simulate Louise's teaching
print("\n" + "="*70)
print("  SIMULATING ZERO-SHOT VOCABULARY ACQUISITION")
print("="*70 + "\n")

xdu = XenolinguisticDecryptionUnit()

pairs = [("human", "◯⟨⟩"), ("Louise", "◯⟨L⟩"), ("Ian", "◯⟨I⟩"),
         ("question", "◯⟨?⟩"), ("purpose", "◯⟨P⟩"), ("weapon", "◯⟨W⟩")]

for h, a in pairs:
    xdu.teaching_session(h, a)

print(f"\n[SYSTEM] Vocabulary size: {len(xdu.vocabulary)} terms")
print("[SYSTEM] This is the BOOTSTRAPPING problem in low-resource NLP!")
````

### 2.2 CLIP Multimodal Grounding

````
# CLIP MULTIMODAL GROUNDING

from transformers import CLIPProcessor, CLIPModel
from PIL import Image, ImageDraw, ImageFont

print("Loading CLIP model...")
model = CLIPModel.from_pretrained('openai/clip-vit-base-patch32')
processor = CLIPProcessor.from_pretrained('openai/clip-vit-base-patch32')
print("[OK] CLIP loaded.\n")

# Create whiteboard
img = Image.new('RGB', (400, 300), 'white')
draw = ImageDraw.Draw(img)
try:
    font = ImageFont.truetype('/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf', 50)
except:
    font = ImageFont.load_default()
draw.text((110, 120), 'HUMAN', fill='black', font=font)

candidates = ['a human being', 'a linguistic symbol', 'a weapon', 'an equation']
inputs = processor(text=candidates, images=img, return_tensors='pt', padding=True)
outputs = model(**inputs)
probs = outputs.logits_per_image.softmax(dim=1)[0]

print("CLIP Interpretation of Whiteboard:\n")
for txt, p in sorted(zip(candidates, probs.detach().numpy()), key=lambda x: x[1], reverse=True):
    print(f"  {p:.1%} {'█'*int(p*30)} {txt}")

plt.figure(figsize=(5,4))
plt.imshow(img)
plt.axis('off')
plt.title("Louise's Whiteboard", color='#00d9ff')
plt.show()
````

> **Key Finding:** Multimodal models like CLIP align visual and textual representations in shared embedding space. Louise's whiteboard method mirrors this: pairing a visual referent (her body) with text ("HUMAN") creates a grounded concept. This is how zero-resource language learning begins: **grounding before grammar**.

### 3.2 Heptapod Logogram Generator

````
# LOGOGRAM GENERATOR

def generate_logogram(concept, entities=5, importance=1.0):
    fig, ax = plt.subplots(figsize=(6, 6), facecolor='black')
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')

    # Main ring
    theta = np.linspace(0.05, 2*np.pi-0.05, 200)
    t = 0.06 * importance
    ax.fill(np.concatenate([(1+t)*np.cos(theta), (1-t)*np.cos(theta)[::-1]]),
            np.concatenate([(1+t)*np.sin(theta), (1-t)*np.sin(theta)[::-1]]), color='white', alpha=0.9)

    # Radial strokes
    for i in range(entities):
        a = 2*np.pi*i/entities
        ax.plot([(1-t)*np.cos(a), 0.5*np.cos(a)], [(1-t)*np.sin(a), 0.5*np.sin(a)],
                color='white', linewidth=3*importance, alpha=0.8)

    plt.title(f'\"{concept.upper()}\"', fontsize=14, color='#00d9ff')
    plt.show()

generate_logogram('Offer Tool', entities=7, importance=1.0)
generate_logogram('Offer Weapon', entities=7, importance=1.8)

print("Same structure, different weight. Observers project bias onto intensity.")
````

> **Key Finding:** The logograms for "Offer Tool" and "Offer Weapon" have identical structure but different stroke weights. Human observers project their cultural bias onto this visual intensity. This mirrors real translation failures where **ambiguous signals are interpreted through the lens of the observer's training context**.

# SESSION 04: TRANSFORMER SELF-ATTENTION

```
┌──────────────────────────────────────────────────────────────────────┐
│ TIMESTAMP: 2016-11-17 03:44:52 GMT                                   │
│ OBJECTIVE: VISUALIZE NON-LINEAR TOKEN RELATIONSHIPS                  │
│ STATUS: THEORETICAL BREAKTHROUGH                                     │
└──────────────────────────────────────────────────────────────────────┘
```

**RNNs** process tokens sequentially (human-like). **Transformers** process all tokens simultaneously (Heptapod-like).

Self-attention allows every token to "see" every other token: non-linear perception.

---

### 4.1 Self-Attention Heatmap

````
# ATTENTION VISUALIZATION

from transformers import BertTokenizer, BertModel
import torch

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert = BertModel.from_pretrained('bert-base-uncased', output_attentions=True)

sentence = 'There is no linear time in the universe'
inputs = tokenizer(sentence, return_tensors='pt')
tokens = tokenizer.convert_ids_to_tokens(inputs['input_ids'][0])

with torch.no_grad():
    outputs = bert(**inputs)

attn = outputs.attentions[-1][0, 0].numpy()  # Last layer, first head

plt.figure(figsize=(10, 8))
sns.heatmap(attn, xticklabels=tokens, yticklabels=tokens, cmap='plasma',
            cbar_kws={'label': 'Attention Weight'})
plt.title('Self-Attention: Non-Linear Token Relationships', fontsize=14, color='#00d9ff')
plt.xlabel('Attending To')
plt.ylabel('Attending From')
plt.show()

print("Every token attends to every other simultaneously.")
print("This is Heptapod-style perception: holistic, not sequential.")
````

### 4.2 Circular Attention Visualization

````
# CIRCULAR ATTENTION (HEPTAPOD STYLE)

fig, ax = plt.subplots(figsize=(10, 10), facecolor='black', subplot_kw=dict(projection='polar'))

num_tokens = len(tokens)
angles = np.linspace(0, 2*np.pi, num_tokens, endpoint=False)

# Place tokens around circle
for i, (ang, tok) in enumerate(zip(angles, tokens)):
    ax.text(ang, 1.15, tok, ha='center', va='center', fontsize=11, color='white',
            bbox=dict(boxstyle='round,pad=0.4', facecolor='#00d9ff', alpha=0.7))

# Draw attention connections
for i in range(num_tokens):
    for j in range(num_tokens):
        if i != j and attn[i,j] > 0.1:
            ax.plot([angles[i], angles[j]], [1, 1], color='white', alpha=attn[i,j], linewidth=attn[i,j]*3)

ax.set_ylim(0, 1.3)
ax.set_yticks([])
ax.set_xticks([])
ax.spines['polar'].set_visible(False)
ax.grid(False)
ax.set_facecolor('black')
plt.title('Heptapod Attention: Holistic Perception', fontsize=16, color='#00d9ff', pad=30)
plt.show()
````

> **Key Finding:** Transformer self-attention allows every token to "see" every other token simultaneously: unlike RNNs which process sequentially. This non-linear perception mirrors Heptapod B's circular structure. The architecture itself encodes a different cognitive relationship with sequence and time.

- **Strong form (film)**: Learning Heptapod B CAUSES non-linear time perception: **Weak form (reality)**: Language INFLUENCES thought patterns

Let's analyze how BERT embeds different temporal concepts.

---

### 5.1 Temporal Concept Embeddings (BERT)

````
# SAPIR-WHORF: EMBEDDING ANALYSIS

from transformers import AutoTokenizer, AutoModel
from sklearn.metrics.pairwise import cosine_similarity

tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
model = AutoModel.from_pretrained('bert-base-uncased')

def get_embedding(text):
    inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True)
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state[:, 0, :].numpy().flatten()

time_concepts = [
    'The event happened in the past',
    'The event happens in the present',
    'The event will happen in the future',
    'All moments exist simultaneously'  # Heptapod view
]

embeddings = np.array([get_embedding(t) for t in time_concepts])
sim_matrix = cosine_similarity(embeddings)

labels = ['PAST', 'PRESENT', 'FUTURE', 'SIMULTANEOUS']

plt.figure(figsize=(8, 6))
sns.heatmap(sim_matrix, annot=True, fmt='.3f', cmap='YlOrRd',
            xticklabels=labels, yticklabels=labels)
plt.title('Temporal Concept Similarity in BERT', fontsize=14, color='#00d9ff')
plt.show()

print("\nNotice: 'SIMULTANEOUS' clusters differently: it breaks linear time!")
print("This mirrors Louise's cognitive shift after learning Heptapod B.")
````

> **Key Finding:** BERT encodes "simultaneous time" differently from linear past/present/future. The embedding distances show that our language models carry implicit temporal assumptions. Learning a new linguistic framework (like Heptapod B) could theoretically shift these representations: a computational echo of the Sapir-Whorf hypothesis.
